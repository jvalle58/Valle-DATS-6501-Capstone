```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import the inc and vote_groups data frames
inc = pd.read_excel('inc_10-08.xlsx')
vote_groups = pd.read_excel('vote_groups_09-30.xlsx')
inc.head()
```

```python
vote_groups.head()
```

```python
#Drop the unnamed column
inc = inc.drop(['Unnamed: 0'], axis=1)
vote_groups = vote_groups.drop(['Unnamed: 0'], axis=1)
inc.head()
```

```python
#Slice the rows pertaining to per-capita income in inc, and calculate the average for each row
inc = inc.loc[(inc.Description.str.contains('Per'))]
y = inc.iloc[:,5:18].reset_index(drop=True)
y['Average Income'] = y.mean(axis=1)
y
```

```python
#Concatenate the last column in y to vote_groups to create a new dataframe
v = pd.concat([vote_groups,y['Average Income']], axis=1)
v.head()
```

```python
#Count the number of times each party preference is recorded
v['PartyPref'].value_counts()
```
