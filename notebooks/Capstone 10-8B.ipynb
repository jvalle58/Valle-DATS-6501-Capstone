```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import the data frames created from the previous Jupyter Notebook.  In this case, we will be merging all of the data
#together in one by year to make it easier for our machine learning analysis
vote = pd.read_excel('vote_10-05.xlsx')
inc = pd.read_excel('inc_10-08.xlsx')
gdp = pd.read_excel('gdp_10-08.xlsx')
vote_groups = pd.read_excel('vote_groups_09-30.xlsx')
vote.head()
```

```python
#Drop the columns listed below, as only the election years will be included later
vote = vote.drop(['Unnamed: 0'], axis=1)
inc = inc.drop(['Unnamed: 0','2005','2006','2007','2009','2010','2011','2013','2014','2015'], axis=1)
gdp = gdp.drop(['Unnamed: 0','2005','2006','2007','2009','2010','2011','2013','2014','2015',
                '2005-Share','2006-Share','2007-Share','2009-Share','2010-Share',
                '2011-Share','2013-Share','2014-Share','2015-Share'], axis=1)
vote_groups = vote_groups.drop(['Unnamed: 0'], axis=1)
vote.head()
```

```python
#Split vote by creating a new data frame for each year
vote4 = vote[vote.Year == 2004].reset_index(drop=True)
vote8 = vote[vote.Year == 2008].reset_index(drop=True)
vote12 = vote[vote.Year == 2012].reset_index(drop=True)
vote16 = vote[vote.Year == 2016].reset_index(drop=True)
vote16.head()
```

```python
#For each year, split inc by its population or per-capita income figures
inc4 = inc.iloc[:,[0,1,2,3,5]]
inc4_pop = inc4.loc[inc4.Description.str.contains('Population')].reset_index(drop=True)
inc4_inc = inc4.loc[inc4.Description.str.contains('Per')].reset_index(drop=True)

inc8 = inc.iloc[:,[0,1,2,3,6]]
inc8_pop = inc8.loc[inc4.Description.str.contains('Population')].reset_index(drop=True)
inc8_inc = inc8.loc[inc4.Description.str.contains('Per')].reset_index(drop=True)

inc12 = inc.iloc[:,[0,1,2,3,7]]
inc12_pop = inc12.loc[inc4.Description.str.contains('Population')].reset_index(drop=True)
inc12_inc = inc12.loc[inc4.Description.str.contains('Per')].reset_index(drop=True)

inc16 = inc.iloc[:,[0,1,2,3,8]]
inc16_pop = inc16.loc[inc4.Description.str.contains('Population')].reset_index(drop=True)
inc16_inc = inc16.loc[inc4.Description.str.contains('Per')].reset_index(drop=True)

inc16_pop.head()
```

```python
#Rename the year column to the proper variable and drop the Description column
inc4_pop = inc4_pop.rename(columns = {'2004':'Population'})
inc4_pop = inc4_pop.drop(['Description'], axis=1)

inc8_pop = inc8_pop.rename(columns = {'2008':'Population'})
inc8_pop = inc8_pop.drop(['Description'], axis=1)

inc12_pop = inc12_pop.rename(columns = {'2012':'Population'})
inc12_pop = inc12_pop.drop(['Description'], axis=1)

inc16_pop = inc16_pop.rename(columns = {'2016':'Population'})
inc16_pop = inc16_pop.drop(['Description'], axis=1)

inc4_inc = inc4_inc.rename(columns = {'2004':'Per-Capita Income'})
inc4_inc = inc4_inc.drop(['Description'], axis=1)

inc8_inc = inc8_inc.rename(columns = {'2008':'Per-Capita Income'})
inc8_inc = inc8_inc.drop(['Description'], axis=1)

inc12_inc = inc12_inc.rename(columns = {'2012':'Per-Capita Income'})
inc12_inc = inc12_inc.drop(['Description'], axis=1)

inc16_inc = inc16_inc.rename(columns = {'2016':'Per-Capita Income'})
inc16_inc = inc16_inc.drop(['Description'], axis=1)

inc16_inc.head()
```

```python
#For each year, concatenate its voting, population, and per-capita income data
data4 = pd.concat([vote4, inc4_pop['Population'], inc4_inc['Per-Capita Income']], axis=1)
data8 = pd.concat([vote8, inc8_pop['Population'], inc8_inc['Per-Capita Income']], axis=1)
data12 = pd.concat([vote12, inc12_pop['Population'], inc12_inc['Per-Capita Income']], axis=1)
data16 = pd.concat([vote16, inc16_pop['Population'], inc16_inc['Per-Capita Income']], axis=1)
data16.head()
```

```python
#For each year, split gdp by its GDP figure and share over the industries
gdp4 = gdp.iloc[:,[0,1,2,3,5,9]]
gdp8 = gdp.iloc[:,[0,1,2,3,6,10]]
gdp12 = gdp.iloc[:,[0,1,2,3,7,11]]
gdp16 = gdp.iloc[:,[0,1,2,3,8,12]]
gdp16.head()
```

```python
#In this cell and the one after the next, convert the industries into columns for each year, and calculate their GDP
#shares in several new ones.  As a convenience for later, reset the index numbers as well
x4 = gdp4.pivot_table(values='2004', index='GeoName', columns='Description', aggfunc='first')
x4['Construction Share'] = (x4['Construction']/x4['Total GDP'])*100
x4['Education and Health Share'] = (x4['Education and Health']/x4['Total GDP'])*100
x4['Finance Share'] = (x4['Finance']/x4['Total GDP'])*100
x4['Food and Recreation Share'] = (x4['Food and Recreation']/x4['Total GDP'])*100
x4['Government Share'] = (x4['Government']/x4['Total GDP'])*100
x4['Information Share'] = (x4['Information']/x4['Total GDP'])*100
x4['Manufacturing Share'] = (x4['Manufacturing']/x4['Total GDP'])*100
x4['Natural Resources Share'] = (x4['Natural Resources']/x4['Total GDP'])*100
x4['Other Share'] = (x4['Other']/x4['Total GDP'])*100
x4['Professional Share'] = (x4['Professional']/x4['Total GDP'])*100
x4['Trade Share'] = (x4['Trade']/x4['Total GDP'])*100
x4['Transportation Share'] = (x4['Transportation']/x4['Total GDP'])*100
x4['Utilities Share'] = (x4['Utilities']/x4['Total GDP'])*100
x4 = x4.reset_index(drop=False)
x4.head()
```

```python
#This is a check to make sure the industry conversion to columns has worked and the industry shares are equivalent to
#the previous data frame corresponding with the same year
gdp4.loc[(gdp4.GeoName == 'Ada, ID'), ['GeoFIPS','GeoName','Description',
                                       '2004','2004-Share']].sort_values(by='Description')
```

```python
x8 = gdp8.pivot_table(values='2008', index='GeoName', columns='Description', aggfunc='first')
x8['Construction Share'] = (x8['Construction']/x8['Total GDP'])*100
x8['Education and Health Share'] = (x8['Education and Health']/x8['Total GDP'])*100
x8['Finance Share'] = (x8['Finance']/x8['Total GDP'])*100
x8['Food and Recreation Share'] = (x8['Food and Recreation']/x8['Total GDP'])*100
x8['Government Share'] = (x8['Government']/x8['Total GDP'])*100
x8['Information Share'] = (x8['Information']/x8['Total GDP'])*100
x8['Manufacturing Share'] = (x8['Manufacturing']/x8['Total GDP'])*100
x8['Natural Resources Share'] = (x8['Natural Resources']/x8['Total GDP'])*100
x8['Other Share'] = (x8['Other']/x8['Total GDP'])*100
x8['Professional Share'] = (x8['Professional']/x8['Total GDP'])*100
x8['Trade Share'] = (x8['Trade']/x8['Total GDP'])*100
x8['Transportation Share'] = (x8['Transportation']/x8['Total GDP'])*100
x8['Utilities Share'] = (x8['Utilities']/x8['Total GDP'])*100
x8 = x8.reset_index(drop=False)

x12 = gdp12.pivot_table(values='2012', index='GeoName', columns='Description', aggfunc='first')
x12['Construction Share'] = (x12['Construction']/x12['Total GDP'])*100
x12['Education and Health Share'] = (x12['Education and Health']/x12['Total GDP'])*100
x12['Finance Share'] = (x12['Finance']/x12['Total GDP'])*100
x12['Food and Recreation Share'] = (x12['Food and Recreation']/x12['Total GDP'])*100
x12['Government Share'] = (x12['Government']/x12['Total GDP'])*100
x12['Information Share'] = (x12['Information']/x12['Total GDP'])*100
x12['Manufacturing Share'] = (x12['Manufacturing']/x12['Total GDP'])*100
x12['Natural Resources Share'] = (x12['Natural Resources']/x12['Total GDP'])*100
x12['Other Share'] = (x12['Other']/x12['Total GDP'])*100
x12['Professional Share'] = (x12['Professional']/x12['Total GDP'])*100
x12['Trade Share'] = (x12['Trade']/x12['Total GDP'])*100
x12['Transportation Share'] = (x12['Transportation']/x12['Total GDP'])*100
x12['Utilities Share'] = (x12['Utilities']/x12['Total GDP'])*100
x12 = x12.reset_index(drop=False)

x16 = gdp16.pivot_table(values='2016', index='GeoName', columns='Description', aggfunc='first')
x16['Construction Share'] = (x16['Construction']/x16['Total GDP'])*100
x16['Education and Health Share'] = (x16['Education and Health']/x16['Total GDP'])*100
x16['Finance Share'] = (x16['Finance']/x16['Total GDP'])*100
x16['Food and Recreation Share'] = (x16['Food and Recreation']/x16['Total GDP'])*100
x16['Government Share'] = (x16['Government']/x16['Total GDP'])*100
x16['Information Share'] = (x16['Information']/x16['Total GDP'])*100
x16['Manufacturing Share'] = (x16['Manufacturing']/x16['Total GDP'])*100
x16['Natural Resources Share'] = (x16['Natural Resources']/x16['Total GDP'])*100
x16['Other Share'] = (x16['Other']/x16['Total GDP'])*100
x16['Professional Share'] = (x16['Professional']/x16['Total GDP'])*100
x16['Trade Share'] = (x16['Trade']/x16['Total GDP'])*100
x16['Transportation Share'] = (x16['Transportation']/x16['Total GDP'])*100
x16['Utilities Share'] = (x16['Utilities']/x16['Total GDP'])*100
x16 = x16.reset_index(drop=False)
```

```python
#Drop the unneeded axis name
x4 = x4.rename_axis(None, axis=1)
x8 = x8.rename_axis(None, axis=1)
x12 = x12.rename_axis(None, axis=1)
x16 = x16.rename_axis(None, axis=1)
x4.head()
```

```python
#In the voting, population, and per-capita income data, sort the names under the Area column in alphabetical order
data4 = data4.sort_values(by='Area').reset_index(drop=True)
data8 = data8.sort_values(by='Area').reset_index(drop=True)
data12 = data12.sort_values(by='Area').reset_index(drop=True)
data16 = data16.sort_values(by='Area').reset_index(drop=True)
data16.head()
```

```python
#Concatenate the GDP data with the voting, population, and income data by year, and sort the resulting data frame in 
#alphabetical order with respect to State and then Area
data4 = pd.concat([data4, x4], axis=1)
data8 = pd.concat([data8, x8], axis=1)
data12 = pd.concat([data12, x12], axis=1)
data16 = pd.concat([data16, x16], axis=1)

data4 = data4.sort_values(by=['State','Area']).reset_index(drop=True)
data8 = data8.sort_values(by=['State','Area']).reset_index(drop=True)
data12 = data12.sort_values(by=['State','Area']).reset_index(drop=True)
data16 = data16.sort_values(by=['State','Area']).reset_index(drop=True)
data16.head()
```
